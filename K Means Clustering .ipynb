### Unsupervised Learning
K Means Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# here we are importing bolb to create a dummy dataset

from sklearn.datasets import make_blobs
```

```python
# creating a dummy dataset using blob
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0].shape
```

```python
sns.set_style('whitegrid')
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey = True, figsize=(12,7))
ax1.set_title('KMeans')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(col.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
def converter(private):
    if private == 'Yes':
        return 1
    return 0
    
```

```python
col['Cluster'] = col['Private'].apply(converter)
```

```python
col.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(col['Cluster'],kmeans.labels_))
print(classification_report(col['Cluster'],kmeans.labels_))
```
